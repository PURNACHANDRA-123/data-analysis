```python
#matrix operation
import numpy as np
```

```python
mat1=np.array([[1,0,3],[4,5,6],[7,0,9]])
mat2=np.array([[1,9,3],[4,9,6],[1,8,9]])
mat3=np.add(mat1,mat2)
mat3
```

```python
mat4=np.subtract(mat1,mat2)
mat4
```

```python

mat5 = np.array([[2, 4, 5], [3, 6, 8], [1, 1, 1], [9, 9, 9], [2, 2, 2]])  # 5x3
mat6 = np.array([[2, 4, 5], [3, 6, 8], [1, 1, 1]])  # 3x3

mat7 = np.dot(mat5, mat6)  # 5x3

print(mat7)
```

```python
#multiplication  #shold have same dimensions
mat8=np.multiply(mat1,mat2)
mat8
```
