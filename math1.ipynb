```python
#ADDICTION OF 2 VECTORS
v1=np.asarray([0,0,2,4])
v2=np.asarray([0,0,-4,-2])
sum=v1+v2
print(sum)
difference=v1-v2
print(difference)
multiply=v1*v2
print(multiply)
multiply1=3*v1
print(multiply1)
multiply2=3*v2
print(multiply2)

```

```python
#PLOTING A VECTOR
plt.quiver(0,1,4,5)
plt.show()
```

```python
plt.quiver(0,0,9,9, scale_units='xy',angles='xy',scale=1,color='r') #1
plt.quiver(0,0,-7,-8, scale_units='xy',angles='xy',scale=1,color='g') #2
#sum of above 2 vectors
plt.quiver(0,0,2,1, scale_units='xy',angles='xy',scale=1,color='b')
#subction of above 2 vectors
plt.quiver(0,0,16,17, scale_units='xy',angles='xy',scale=1,color='y')


plt.xlim(-10,20)
plt.ylim(-10,20)

plt.show()
```
