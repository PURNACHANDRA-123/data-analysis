```python
#used for making plots
import matplotlib.pyplot as plt
import numpy as np # Make sure to import numpy and alias it as np


```

```python
x=np.linspace(0,10,4)
print(x)
```

```python
y=np.sin(x)
z=np.cos(x)
print(y)
print(z)
```

```python
#ploting the data
plt.plot(y,z)
plt.show()
```

```python
#sine wave
plt.plot(x,y)
plt.show()
#cose wave
plt.plot(x,z)
plt.show()
```

```python
#adding title , xaxis & y-axis labels
plt.plot(x,y, color='blue', marker='.', linestyle='--', markersize=10)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
#adding title , xaxis & y-axis labels
plt.plot(x,z, color='red', marker='.', linestyle='--', markersize=8)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cosine Wave')
plt.show()
```

```python
a=np.linspace(-20,20,30)
plt.plot(a,a**2,'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Equation')
plt.show()
```

```python
#printing with green colour
plt.plot(a,a**2,'b.')
plt.show()
```

```python
c=np.linspace(-10,10,100)
plt.plot(c,c**2,'r--')
plt.show()
plt.plot(c,np.sin(c),'y--')

plt.plot(c,np.cos(c),'g--')
plt.show()
```

```python
#BAR PLOT
FIG=plt.figure()
ax=FIG.add_axes([0,0,1.9,1])
langs=['C','C++','Java','Python','PHP','javascript']
students=[23,17,35,45,12,29]
ax.bar(langs,students)
plt.xlabel('Programming Languages')
plt.ylabel('Number of Students')
plt.title('Bar Plot')
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,40)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x, y, c='g', label='sin(x)')  # Scatter points in green
plt.plot(x, y, 'g--')  # Connect points with a green dashed line

# Scatter plot for the cosine wave
plt.scatter(x, z, c='b', label='cos(x)')  # Scatter points in blue
plt.plot(x, z, 'b--')  # Connect points with a blue dashed line

plt.legend()  # Display the legend
plt.show()
```

```python
#3D scatter plot
fig2=plt.figure()
ax=plt.axes(projection='3d')
x1=20*np.random.random(100)
y1=np.sin(x1)
z1=np.cos(x1)
ax.scatter3D(x1,y1,z1,color='blue')
plt.show()
```
