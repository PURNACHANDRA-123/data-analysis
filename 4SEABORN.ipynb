```python
#seaborn #data visualization library
import seaborn as sns
#matplotlib #plotting library
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#seavorn has build-in datasets
#TOTAL BILL VS TIP DATASET
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting theme for the plot
sns.set_theme(style='darkgrid')
```

```python
#visualize the data
sns.relplot(data=tips,x = 'total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#house price dtatsets
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
housing_df=pd.DataFrame(housing.data,columns=housing.feature_names)
housing_df['MedHouseVal']=housing.target
housing_df.head()
```

```python
sns.displot(housing_df['MedHouseVal'])
#distrbution of the plot
```

```python
#+correaltion -correaltion
#heat map
correlation = housing_df.corr() # Fix: Corrected the spelling of 'correlation'

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues') # Fix: using the correctly spelled 'correlation'
```
